## 請搜尋目標為多標籤問題 (Multi-label) 的機器學習案例，了解其資料來源、目標以及評估指標為何
## Hint: 服飾

    英雄聯盟(league of legends)在角色定位上多為多元分類，但部分角色為多標籤。
    主要分類有：
              開戰型坦克（Vanguards）、護排型坦克（Wardens）、
              坦型鬥士（Juggernauts）、切陣型鬥士（Divers）
              刺客（Assassins）、單兵（Skirmishers）,	
              爆發型法師（Burst Mages）、戰鬥型法師（Battle Mages）、長距離法師（Artillery Mages）
              支援型術士（Enchanters）、干擾型術師（Disruptors）
              射手（Marksmen）
    多標籤角色舉例：
              特朗德：護排型坦克兼坦型鬥士
              法洛士：射手兼長距離法師
              凱爾：射手兼支援型術士
              哎可尚：射手兼刺客
    機器學習案例：網站opgg經過玩家多場遊戲，在遊戲開始前就能計算出哪個定位的角色勝率較高。
    目標：讓對遊戲還不熟悉的玩家們，能透過網站對各角色的分類與理解，獲得更佳的遊戲體驗。
    評估指標：玩家在賽季中所有對戰的平均勝率
    資料來源：英雄聯盟官網
